```python
import nltk               # import libraries with there functions
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import spacy

tweets = [ # Sample dataset: tweets
    "I am loving Masters in Artificial Intelligence It's amazing!",
    "Running late for Assignment, ugh need to complete this fast.",
    "The assignment was really exciting and fun."
]

ps = PorterStemmer()  # initialize NLTK tools
lemmatizer = WordNetLemmatizer()

nlp = spacy.load("en_core_web_sm") # initialize SpaCy

def process_nltk(text): # function to process tweets  with NLTK
    tokens = word_tokenize(text)  # tokenization
    stemmed = [ps.stem(token) for token in tokens] #stemming
    lemmatized = [lemmatizer.lemmatize(token) for token in tokens] #lematization
    return tokens, stemmed, lemmatized

# Function to process text with SpaCy
def process_spacy(text):
    doc = nlp(text)
    tokens = [token.text for token in doc] # tokenization
    # lemmatization (SpaCy does not have direct stemming)
    lemmatized = [token.lemma_ for token in doc]
    return tokens, lemmatized

# Process and display results
print("\nSpaCy Processing")
for tweet in tweets:
    tokens, lemmatized = process_spacy(tweet)
    print(f"\nTweet: {tweet}")
    print(f"Tokens: {tokens}")
    print(f"Lemmatized: {lemmatized}")
print("\nNLTK Processing")
for tweet in tweets:
    tokens, stemmed, lemmatized = process_nltk(tweet)
    print(f"\nTweet: {tweet}")
    print(f"Tokens: {tokens}")
    print(f"Stemmed: {stemmed}")
    print(f"Lemmatized: {lemmatized}")

```

**VISUALISATION AND COMPARISON OF NLTK AND SPACY**

```python
import nltk
import spacy
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt

# Sample tweets
tweets = [
    "I am loving Masters in Artificial Intelligence. It's amazing!",
    "Running late for Assignment, ugh need to complete this fast.",
    "The assignment was really exciting and fun."
]

# Initialize SpaCy and download NLTK punkt tokenizer if not already available
nltk.download('punkt')
nlp = spacy.load("en_core_web_sm")

# Combine all tweets into a single text
text = " ".join(tweets)

# Tokenization with NLTK
nltk_tokens = word_tokenize(text)
nltk_freq = Counter(nltk_tokens)

# Tokenization with SpaCy
spacy_doc = nlp(text)
spacy_tokens = [token.text for token in spacy_doc]
spacy_freq = Counter(spacy_tokens)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].bar(*zip(*nltk_freq.most_common(10)), color='red')
axs[0].set_title("NLTK Token Frequency")
axs[0].tick_params(axis='x', rotation=45)

axs[1].bar(*zip(*spacy_freq.most_common(10)), color='green')
axs[1].set_title("SpaCy Token Frequency")
axs[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**CODE FOR NER (NAMED ENTITY RECOGNITION) **

```python
from transformers import pipeline

# Initialize BERT-based NER
pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", tokenizer="dslim/bert-base-NER")

sentences = [
    "My name is Ashish Sikarvar.",
    "I am currently living in Berlin, capital of Germany.",
    "I study Masters in Artificial Intelligence.",
    "Berlin School of Bussiness and Innovation is my University.",
    "Currently, I am preparing an Final Submission Assigment.",
    "The subject is Natural Language Processing.",
    "Name of my Professor is Ahmed Hassan.",
    "This code is for BERT model using hugging face.",
    "This model is used to find the classify and identify entities.",
    "Finally I have reached last sentence ughhh."
]

# NER and display results
for idx, sentence in enumerate(sentences, 1):  #idx means from sentence 1 to last (10)
    entities = pipeline(sentence)
    print(f"\nSentence {idx}: {sentence}")
    print("Entities:")
    for entity in entities:
        print(f"  - {entity['word']} ({entity['entity']}, score: {entity['score']:.2f})")
```

**WORD EMBEDING AND MODEL TRAINING**

```python
# Import Word2Vec from gensim and word tokenizer from nltk

from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk

nltk.download('all')  # download all packages defined in nltk

# Data Corpus
SampleCorpus = [
    "Isreal attacked on Iran this Friday",
    "In retaliation to that Iran Fired around 250 Ballastic missiles on Satureday.",
    "Maximum leadership of IRAN has been eliminated.",
    "Isreal also suffered from attacks from Iran and many people died in Tel-Aviv.",
    "I reached a point where I have to stop writing and run this code."
]

# Break sentence into words from SampleCorpus
Tok_text = [
    [word.lower() for word in word_tokenize(sentence) if word.isalpha()]
    for sentence in SampleCorpus
]

# Train Word2Vec model
w2v_model = Word2Vec(sentences=Tok_text, vector_size=60, window=5, min_count=1)

# Print the vector for the word in sentence
print("vector form of Iran'-->\n", w2v_model.wv['iran'])

# Print the similar word to a word in sentence
print("\n\nWords similar to 'from':")
for word, score in w2v_model.wv.most_similar('from'):
    print(f"{word}: {score:.2f}")
```

**GETTING COSINE SIMILARITY OF THE TEXT**

```python
# function to get cosine similarity for word 'iran'
cosine_similarity = w2v_model.wv.most_similar('iran', topn=10)

print("ten words similar to 'iran':\n")
for i, Simil in cosine_similarity:
    print(f"{i}: {Simil}")  # printing the result
```

**MODEL TRAINING THROUGH LOGISTIC REGRESSION**

```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Load the CSV file (upload in Google Colab)
df = pd.read_csv('/content/emotion-labels-test.csv')

# Clean the text data
def clean_text(text):
    text = ' '.join(text.split())  # Remove extra whitespace
    text = re.sub(r'http\S+|www\S+|[^\w\s]', '', text)  # Remove URLs and punctuation
    text = text.lower()  # Convert to lowercase
    return text

# Remove rows with missing data
df = df.dropna(subset=['text', 'label'])

# Apply preprocessing
df['text'] = df['text'].apply(clean_text)
df['label'] = df['label'].str.lower()

# Split into features and target
X = df['text']
y = df['label']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)

# Define pipeline TfidfVectorizer
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english', max_features=500)),
    ('logreg', LogisticRegression(max_iter=500))
])

# Train the model
pipeline.fit(X_train, y_train)

# Make predictions
y_pred = pipeline.predict(X_test)

# Calculate and print metrics
Accuracy = accuracy_score(y_test, y_pred)
Precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
Recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1_score = f1_score(y_test, y_pred, average='weighted', zero_division=0)

print(f"Performance Metrics --> \n")
print(f"Accuracy --> {Accuracy}")
print(f"Precision --> {Precision}")
print(f"Recall   --> {Recall}")
print(f"F1-Score --> {f1_score}")

# classification report for each emotions
print("\nDetailed Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))
```

**CODE FOR DISTILBERT MODEL ON EMOTIONS DATASET**

```python
import pandas as pd
import torch
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix # Import confusion_matrix here
import matplotlib.pyplot as plt # Import matplotlib for plotting
import seaborn as sns # Import seaborn for heatmap

#load the dataset
data = pd.read_csv('/content/emotion-labels-test.csv')

#emotions defined in dataset
emo_list = ['joy', 'fear', 'sadness', 'anger']
data = data[data['label'].isin(emo_list)]

# emotion converstion to numbers
label_encoder = LabelEncoder()
data['label_num'] = label_encoder.fit_transform(data['label'])

#data split into testing and trainign
train, test, train_emotion, test_emotion = train_test_split(
    data['text'], data['label_num'], test_size=0.2, random_state=1
)

# DistilBert model and Tokenization of data
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

def tokenize(texts):
    return tokenizer(texts.tolist(), padding=True, truncation=True, max_length=128)

train_encodings = tokenize(train)
test_encodings = tokenize(test)

# Data processing into a class
class EmotionDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = EmotionDataset(train_encodings, train_emotion.tolist())
test_dataset = EmotionDataset(test_encodings, test_emotion.tolist())

# load the pre-trained DistilBERT model
model = DistilBertForSequenceClassification.from_pretrained(
    'distilbert-base-uncased', num_labels=len(label_encoder.classes_)
)

# define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=2,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_strategy="epoch"  # evaluation strategy
)

# creation of trainer and train model
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

trainer.train()

#Evaluation of  the model
results = trainer.predict(test_dataset)
predicted = results.predictions.argmax(axis=1)

print("\nClassification Report:")
print(classification_report(test_emotion, predicted, target_names=label_encoder.classes_))

#Plot confusion matrix
cm = confusion_matrix(test_emotion, predicted)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges',
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Emotion')
plt.ylabel('Actual Emotion')
plt.tight_layout()
plt.show()
```
